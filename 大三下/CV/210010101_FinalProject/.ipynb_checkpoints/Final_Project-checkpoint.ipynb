## 解析图片信息

```python
def get_df_train(annotations_path,images_path):
    # 获取所有的xml文件
    annotations = [os.path.join(directory_path, f) for directory_path, directory_name, 
            files in os.walk(annotations_path) for f in files if f.endswith('.xml')]
    annotation_list = []
    # 对于每个xml文件，对应解析出图片的文件名，类别，高度，宽度，坐标信息
    for annotations_path in annotations:
        root = ET.parse(annotations_path).getroot()
        annotation = {}
        annotation['filename'] = Path(str(images_path) + '/'+ root.find("./filename").text)
        annotation['class'] = root.find("./object/name").text
        annotation['height'] = root.find("./size/height").text
        annotation['width'] = root.find("./size/width").text
        annotation['ymin'] = int(root.find("./object/bndbox/ymin").text)
        annotation['xmin'] = int(root.find("./object/bndbox/xmin").text)
        annotation['ymax'] = int(root.find("./object/bndbox/ymax").text)
        annotation['xmax'] = int(root.find("./object/bndbox/xmax").text)
        annotation_list.append(annotation)
    return pd.DataFrame(annotation_list)
```

## 统一图像和边界框的大小

```python
#创建边界框掩码
def get_mask(bounding_box, image):
    
    height, width = image.shape[0:2]
    mask = np.zeros((height, width), dtype=np.float32)

    # 获取边界框相应坐标
    left, top, right, bottom = map(int, bounding_box)
    
    # 确保边界框坐标位于图像内部
    left, top = max(0, left), max(0, top)
    right, bottom = min(width, right), min(height, bottom)
    
    mask[top:bottom, left:right] = 1.0
    return mask

# 根据掩码获取边界框坐标
def get_bb_by_mask(mask):

    rows, cols = np.nonzero(mask)

    if len(rows) == 0 or len(cols) == 0:
        bounding_box = np.zeros(4, dtype=np.float32)
        return bounding_box
    
    # 获取边界框的各个坐标
    left = np.min(cols)
    top = np.min(rows)
    right = np.max(cols)
    bottom = np.max(rows)
    # 整体坐标
    bounding_box = np.array([left, top, right, bottom], dtype=np.float32)
    
    return bounding_box

# 统一图像和边界框的大小
def resize_image_and_bb(read_path,write_path,bb,size):
    # 读取图像
    image_read = cv2.imread(str(read_path))
    image = cv2.cvtColor(image_read, cv2.COLOR_BGR2RGB)

    # 定义新的大小
    new_size = int(1.45 * size)
    # 调整图像大小
    image_resized = cv2.resize(image, (new_size, size))
    # 获取掩码并调整掩码大小
    mask = get_mask(bb, image)
    mask_resized = cv2.resize(mask, (new_size, size))
    # 根据掩码获取对应的调整后的边界框
    bb_resized = get_bb_by_mask(mask_resized)

    # 获取输出的路径
    output_path = write_path / read_path.name
    # 保存图像
    cv2.imwrite(str(output_path), cv2.cvtColor(image_resized, cv2.COLOR_RGB2BGR))
    
    return output_path, bb_resized
```

```python
new_paths = []
new_bbs = []
resized_path = Path('C:\\Users\\lenovo\\Desktop\\210010101_FinalProject\\resized_images')
# 对于每一行图像都需要统一图像大小和边界框
for index, row in df_train.iterrows():
    # 首先需要获取边界框坐标
    bb_array = np.array([row.values[4],row.values[5],row.values[6],row.values[7]])
    # 调用函数进行调整
    new_path, new_bb = resize_image_and_bb(row['filename'], resized_path, bb_array,250)
    new_paths.append(new_path)
    new_bbs.append(new_bb)
df_train['new_path'] = new_paths
df_train['new_bb'] = new_bbs
```

## 定义数据预处理需要用到的图像变换

```python
# 对图像和掩码进行旋转
def rotate(image, mask, edge_strategy=cv2.BORDER_REFLECT, resampling_method=cv2.INTER_AREA):
    # 随机的旋转角度
    rotation_degree = (np.random.random() - 0.5) * 20
    # 获取旋转中心和旋转变换
    image_height, image_width = image.shape[:2]
    image_center = (image_width / 2, image_height / 2)
    mask_height, mask_width = mask.shape[:2]
    mask_center = (mask_width / 2, mask_height / 2)
    image_rotation = cv2.getRotationMatrix2D(image_center, rotation_degree, scale=1)
    mask_rotation = cv2.getRotationMatrix2D(mask_center, rotation_degree, scale=1)
    
    # 获取随机旋转的图像和掩码
    rotated_image = cv2.warpAffine(image, image_rotation, (image_width, image_height), borderMode=edge_strategy, flags=resampling_method)
    rotated_mask = cv2.warpAffine(mask, mask_rotation, (mask_width, mask_height), borderMode=cv2.BORDER_CONSTANT)
    
    return rotated_image,rotated_mask

# 对图像和掩码进行随机裁剪
def random_crop(image, mask, border_pixels=8):
    image_rows, image_cols = image.shape[:2]
    
    crop_height = image_rows - 2 * border_pixels
    crop_width = round(border_pixels * image_cols / image_rows)
    
    rand_row = random.uniform(0, 1)
    rand_col = random.uniform(0, 1)
    
    start_row = np.floor(2 * rand_row * border_pixels).astype(int)
    start_col = np.floor(2 * rand_col * crop_width).astype(int)

    cropped_image = image[start_row:start_row+crop_height, start_col:start_col+image_cols-2*crop_width]
    cropped_mask = mask[start_row:start_row+crop_height, start_col:start_col+image_cols-2*crop_width]
    return cropped_image, cropped_mask

# 对图像和掩码进行中心裁剪
def center_crop(image, mask,vertical_margin=8):
    image_height, image_width, _ = image.shape
    image_horizontal_margin = round((image_width / image_height) * vertical_margin)
    
    image_start_vertical = vertical_margin
    image_end_vertical = image_height - vertical_margin
    image_start_horizontal = image_horizontal_margin
    image_end_horizontal = image_width - image_horizontal_margin
    
    mask_height, mask_width = mask.shape
    mask_horizontal_margin = round((mask_width / mask_height) * vertical_margin)
    
    mask_start_vertical = vertical_margin
    mask_end_vertical = mask_height - vertical_margin
    mask_start_horizontal = mask_horizontal_margin
    mask_end_horizontal = mask_width - mask_horizontal_margin
    
    center_cropped_image = image[image_start_vertical:image_end_vertical, image_start_horizontal:image_end_horizontal]
    center_cropped_mask = mask[mask_start_vertical:mask_end_vertical, mask_start_horizontal:mask_end_horizontal]

    return center_cropped_image,center_cropped_mask

# 整体变换
def transforms(image_path, bounding_box, apply_transforms):
    # 读取图像
    image_read = cv2.imread(str(image_path)).astype(np.float32)
    image = cv2.cvtColor(image_read, cv2.COLOR_BGR2RGB) / 255.0
    
    # 创建掩码
    mask = get_mask(bounding_box, image)
    
    if apply_transforms:
        # 旋转
        image, mask = rotate(image, mask)
        # 随机裁剪
        image, mask = random_crop(image, mask)
    else:
        # 中心裁剪
        image, mask = center_crop(image, mask)
    
    # 将图像转换为PyTorch张量
    return image, get_bb_by_mask(mask)
```

## 定义数据集类并加载数据集
### 需要使用之前定义的变换对图像进行预处理

```python
class MyDataset(Dataset):
    # 初始化数据集
    def __init__(self, y, paths, bb, apply_transforms=False):
        self.y = y.values if isinstance(y, pd.Series) else y
        self.paths = paths.values if isinstance(paths, pd.Series) else paths
        self.bb = bb.values if isinstance(bb, pd.Series) else bb
        self.apply_transforms = apply_transforms

    # 获取图像数量
    def __len__(self):
        return len(self.paths)
    
    # 根据索引获取相应的图像，标签和边界框
    def __getitem__(self, index):
        y_class = self.y[index]
        path = self.paths[index]
        # 使用之前定义的变换对图像进行预处理
        image, y_bb = transforms(path, self.bb[index], self.apply_transforms)
        # 归一化
        image = (image - [0.485, 0.456, 0.406])/[0.229, 0.224, 0.225]
        image = np.rollaxis(image, 2)
        image = torch.tensor(image, dtype=torch.float32)
        
        return image, y_class, y_bb
```

```python
# 加载数据集并创建dataloader
batch_size = 32
df_train = df_train.reset_index()
X = df_train[['new_path', 'new_bb']]
Y = df_train['class']
X_train, X_val, y_train, y_val = train_test_split(X, Y, test_size=0.2, random_state=35)
dataset_train = MyDataset(y_train,X_train['new_path'],X_train['new_bb'] , apply_transforms=True)
dataset_val = MyDataset(y_val,X_val['new_path'],X_val['new_bb'])
dataloader_train = DataLoader(dataset_train, batch_size=batch_size, shuffle=True)
dataloader_val = DataLoader(dataset_val, batch_size=batch_size)
```

## 构建模型

```python
class MyModel(nn.Module):
    def __init__(self):
        super(MyModel, self).__init__()
        # 使用resnet18作为特征提取器，使用的权重为默认的参数
        resnet = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
        self.layers = nn.Sequential(*list(resnet.children())[:8])
        
        # 分类器
        self.classifier = nn.Sequential(
            nn.BatchNorm1d(512),
            nn.Linear(512, 4)
        )
        
        # 边界框回归器
        self.bb = nn.Sequential(
            nn.BatchNorm1d(512),
            nn.Linear(512, 4)
        )
        
    def forward(self, x):
        # 前向传播
        x = self.layers(x)
        x = F.relu(x)
        x = nn.AdaptiveAvgPool2d((1,1))(x)
        x = x.view(x.shape[0], -1)
        return self.classifier(x), self.bb(x)
```

## 定义训练相关函数

```python
def val_metrics(model, dataloader_val, C=1000):
    model.eval()
    # 初始化
    total_samples = 0
    total_loss = 0.0
    total_correct = 0
    # 对于验证集的每个batch进行评估
    for x, y_class, y_bb in dataloader_val:
        batch = y_class.shape[0]
        x = x.float()
        y_bb = y_bb.float()

        # 获取预测的类别和边界框
        out_class, out_bb = model(x)

        # 类别损失
        loss_class = F.cross_entropy(out_class, y_class, reduction="sum")
        # 边界框坐标损失
        loss_bb = F.l1_loss(out_bb, y_bb, reduction="none").sum(1).sum()
        loss = loss_class + loss_bb/C
        
        #计算准确率
        _, pred = torch.max(out_class, 1)
        total_correct += pred.eq(y_class).sum().item()
        # 计算总样本数和总损失
        total_samples += batch
        total_loss += loss.item()

    return total_loss/total_samples, total_correct/total_samples

# 训练函数
def train(model, optimizer, dataloader_train, val_dl, epochs=10,C=1000):
    # 对于每个epoch
    for epoch in range(epochs):
        model.train()
        # 初始化
        total_samples = 0
        total_loss = 0
        # 对于训练集的每个batch进行训练
        for x, y_class, y_bb in dataloader_train:
            batch = y_class.shape[0]
            x = x.float()
            y_bb = y_bb.float()

            # 获取预测的类别和边界框
            out_class, out_bb = model(x)

            # 计算类别损失
            loss_class = F.cross_entropy(out_class, y_class, reduction="sum")

            # 边界框坐标损失
            loss_bb = F.smooth_l1_loss(out_bb, y_bb, reduction="none").sum(1).sum()
            loss = loss_class + loss_bb/C

            # 梯度清零
            optimizer.zero_grad()
            # 反向传播
            loss.backward()
            # 更新参数
            optimizer.step()
            total_samples += batch
            total_loss += loss.item()
        # 计算训练损失
        train_loss = total_loss/total_samples
        # 计算损失和准确率
        val_loss, val_acc = val_metrics(model, dataloader_val, C)
        print("train loss: %.4f; val loss: %.4f; val acc: %.4f" % (train_loss, val_loss, val_acc))
    return total_loss/total_samples

# 更新优化器的学习率
def update_lr(optimizer, lr):
    for i, param_group in enumerate(optimizer.param_groups):
        param_group["lr"] = lr
```

## 训练模型

```python
# 训练模型
model = MyModel()
parameters = filter(lambda p: p.requires_grad, model.parameters())
optimizer = torch.optim.Adamax(parameters, lr=0.005)
train(model, optimizer, dataloader_train, dataloader_val, epochs=10)
```

```python
# 更新学习率进行训练
update_lr(optimizer, 0.002)
train(model, optimizer, dataloader_train, dataloader_val, epochs=10)
```
