```python
# 本次作业你可能需要用到的包和函数
import numpy as np  
from skimage import io  
from sklearn.decomposition import PCA  
import matplotlib.pyplot as plt

# plt显示灰度图片
def plt_show(img):
    plt.imshow(img,cmap='gray')
    plt.show()

# 读取一个文件夹下的所有图片，输入参数是文件名，返回文件地址列表
def read_directory(directory_name):
    faces_addr = []
    for filename in os.listdir(directory_name):
        faces_addr.append(directory_name + "/" + filename)
    return faces_addr
```

```python
# Your Code Here
# task1
import os
import cv2
origin_imgs = np.empty((5,21,256*256))
reduced_imgs = np.empty((5,21,8))
pcas = [0,0,0,0,0]
for index in range(97, 102):
    sub_dir = "data" + "/" + "1" + chr(index)
    pcas[index-97] = PCA(n_components=8)
    i = 0
    for face_addr in read_directory(sub_dir):
        # print(face_addr)
        face_img = cv2.imread(face_addr, cv2.IMREAD_GRAYSCALE)
        origin_imgs[index-97][i]=face_img.flatten()
        i = i+1
    #print(origin_imgs[index-97].shape)
    reduced_imgs[index-97] = pcas[index-97].fit_transform(origin_imgs[index-97])
```

```python
#task2
eigen_imgs = [[] for _ in range(5)]
def display_img(eigen_images,rows,cols,img_height,img_width):
    fig, axs = plt.subplots(rows, cols, figsize=(15, 6))
    for i in range(rows):
        for j in range(cols):
            idx = i * cols + j
            axs[i, j].imshow(eigen_images[idx].reshape(img_height,img_width), cmap='gray')
            axs[i, j].axis('off')
    plt.show()
for i in range(0,5):
    print("第",i+1,"个子目录,8张对应的eigen-image为：\n")
    eigen_imgs[i] = pcas[i].components_
    display_img(eigen_imgs[i],2,4,256,256)
```

```python
# task3
diffs = [0,0,0,0,0]
b = np.empty((5,21,256*256))
for i in range(0,5):
    c = pcas[i].transform(origin_imgs[i])
    b[i] = pcas[i].inverse_transform(c)
    for index in range(0,21):
        diffs[i] += np.mean((origin_imgs[i][index]-b[i][index])**2)
    diffs[i] = diffs[i]/21
    print("第",i+1,"个子目录,平均重构误差为：",diffs[i])
```

```python
#task4
for i in range(0,5):
    print("第",i+1,"子目录，序号1的重建图像为：\n")
    plt_show(b[i][0].reshape(256,256))
```
